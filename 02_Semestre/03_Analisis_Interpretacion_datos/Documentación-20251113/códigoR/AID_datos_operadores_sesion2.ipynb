# Variables y constantes

El nombre de una variable tiene que ser claro y debe reflejar el dato que contiene, es decir, descriptivo. Se debe tener en cuenta que es un lenguaje sensible a las mayúsculas. Por ejemplo, EDAD y edad son diferentes.


```python
%%R
# Operador de asignación <-
nacionalidad <- "Mexicana"
estatura <- 1.53
# estudiantes_sesion1 <- 27
estudiantes_sesion1 <- as.integer(7)
# estudiantes_sesion1 <- 27L
usa_lentes <- TRUE
```

# Tipos de datos

```python
# Datos NA y NULL
#
# NA representa un dato perdido. Mientras que NULL indica la ausencia de datos.

%%R
print(class(nacionalidad))
print(class(estatura))
print(class(estudiantes_sesion1))
print(class(usa_lentes))
print(class(NULL))
```

### Validación de tipo de datos

```python
%%R
print(is.integer(estudiantes_sesion1))
print(is.numeric(estatura))
print(is.character(nacionalidad))
print(is.logical(usa_lentes))
print(is.na(NA))
print(is.null(NA))
```

## Coerción

Transformación de un tipo de datos a otro.

Cuando se ejecuta una operación, R hace la coerción de manera implicita. El orden de las coerciones es:

**logical -> integer -> numeric -> character**

```python
%%R
print(as.integer(estatura))
print(as.numeric(usa_lentes))
print(as.character(estatura))
# La siguiente línea marca error
# print(as.logical(nacionalidad))
print(as.logical(estudiantes_sesion1))
print(as.null(NA))
```

## Operadores

### Aritméticos

Se pueden usar en datos de tipo *numeric* e *integer*.

*   '+'  Suma
*   '-'  Resta
*   '*'  Multiplicación
*   '/'  División
*   '^'  Potencia
*   '%%' Residuo de la división
*   '%/%' División entera

```python
%%R
print(4 + 5)
print(9 - 8.5)
print(13.2 * 7)
print(354 / 4)
print('Cociente')
print(354 %/% 4)
print('Residuo')
print(354 %% 4)
print(2 ^ 4)
```

### Relacionales

Se emplean para comparaciones y devuelven como resultado **TRUE** o **FALSE**.

* <    Menor que
* <=   Menor o igual que
* '>'    Mayor que
* '>='   Mayor o igual que
* ==   Exactamente igual
* !=   No es igual

Cuando se emplean estos operadores con cadenas de texto, la comparación se hace por orden alfabético.

```python
%%R
print(nacionalidad == "Americana")
print(estudiantes_sesion1 > estatura)
print(usa_lentes <= estatura)
print(estatura != nacionalidad)

```

### Lógicos

Operaciones de Álgebra Booleana.

* x | y  - x o y - x, y o ambos son verdaderos
* x & y  - x y y - ambos son verdaderos
* !x     - negación de x
* isTRUE(x) - x es verdadero

Se emplean con datos de tipo **numérico**, **lógico** y **complejo**. Además el tipo del dato que devuelven es **logic**

La transformación de un dato **numérico** y **complejo** es **TRUE** si tienen un valor diferente de 0 y **FALSE** en caso contrario.

```python
%%R
print(3 & 4)
print(3 & 0)
print(FALSE | usa_lentes)
# Los operadores lógicos no se pueden usar en cadenas de caracteres, por lo tanto la siguente línea marca error
# print(!nacionalidad)
print(!estatura)
print(estatura | FALSE)
```

### Prioridad de operadores

1. ^
2. '*', /
3. '+', -
4.	<, >, <=, >=, ==, !=
5.	!
6.	&
7.	|
8.	<-

En caso de que se quiera cambiar el orden de ejecucición se deben usar paréntesis.

```python
%%R
print(3 + 2 ^ 3 * 7 %% 6)
print(3 + (2 ^ 3) * 7 %% 6)
print(3 + ((2 ^ 3) * (7 %% 6)))
```
