# Estructuras de datos

Las estructuras de datos son objetos que contienen datos, lo que las distingue es su *número de dimensiones* y si son *homogeneas* o *heterogeneas*.

Las estructuras homogeneas son:
1. Vector
2. Matriz
3. Array

Las estructuras heterogeneas son:
1. Lista
2. Dataframe


```python
%%R
print(is.vector(estudiantes_sesion1))
print(length(estudiantes_sesion1))

print(is.vector(nacionalidad))
print(length(nacionalidad))
```

## Secuencias

Para generar una secuencia se emplea el operador **:**

```python
%%R
print(-10:10)
print(20:10)
print(2.5:15.5)
# Se redondean las cifras del segundo valor
print(2:15.5)
```

Generación de un vector mediante una secuencia

```python
%%R
vec_num <- c(1:10)
vec_num
```

## Operaciones aritmeticas con vectores

```python
%%R
vec_num_neg = vec_num * -1
vec_num_coc = vec_num %/% 2
vec_num_add = vec_num + 10
vec_num_less = vec_num <= 5
print(vec_num_neg)
print(vec_num_coc)
print(vec_num_add)
print(vec_num_less)
```

# Matrices y arrays

Estas estructuras de datos son vectores multidimensionales. En el caso de las matrices tienen **dos dimensiones**, largo y alto, mientras que un array tiene un número arbitrario de dimensiones.

## Matrices

### Creación

Se emplea la función **matrix()**, la cual acepta 2 argumentos *nrow* y *ncol*

```python
%%R

# Los datos se acomodan de arriba hacia abajo y de izquierda a derecha, hasta completar el número de filas y columnas
print(matrix(1:10))
# Cinco renglones
print(matrix(1:10, nrow=5))
# Tres columnas. Los elementos se reciclan
print(matrix(1:10, ncol=3))
# Tres renglones y tres columnas. Se omiten los valores que ya no se pueden acomodar
print(matrix(1:10, nrow=3, ncol=3))
```

### Unión de vectores para crear una matriz

Se usan las funciones
* *cbind()* Une los vectores como columnas
* *rbind()* Une los vectores como filas

```python
%%R
mat_cols = cbind(vec_num_add, vec_num_coc, vec_num_less, vec_num_neg)
mat_rows = rbind(vec_num_less, vec_num_add, vec_num_neg, vec_num_coc)
print(mat_cols)
print(mat_rows)
```

```python
%%R
# dim es exclusivo de matrices o arrays. En caso de usarse en un objeto unidimensional da NULL
print(dim(mat_cols))
print(dim(vec_edades))
```

### Operaciones con matrices

```python
%%R

# Para calcular la transpuesta de una matriz se emplea la función (t)
mat_cols_t = t(mat_cols)
mat_cols_t
```

```python
%%R
print(mat_cols_t + mat_rows)
print(mat_cols * 2)
print(mat_rows %% 5)
```

# Dataframes

Son estructuras de datos de dos dimensiones que pueden contener diferentes tipos de datos, es decir, son heterogéneas. Es la estructura más usada para el análisis de datos.

Los datos de una columna son de un mismo tipo, pues representan atributos, rasgos o variables. Por otro lado, los datos de un reglon pueden ser de diferentes tipos, ya que representan la información de un individuo u observaciones.

Se pueden realizar operaciones aritméticas sólo en sus columnas numéricas, en el resto asigna NA

NOTA: Se debe tener presente que un dataframe esta compuesto por vectores.

## Creación de dataframes

Se usa la función **data.frame()**

```python
%%R
df_estudiantes <- data.frame(
   "nacionalidd" = c("Mexicana", "Guatemalteca", "Venezolana", "Dominicana"),
   "edad" = c(32, 29, 48, 40),
   "sexo" = c("M", "F", "M", "F")
)

df_estudiantes
```

```python
%%R
print(dim(df_estudiantes))
# Numero de columnas
print(length(df_estudiantes))
# Nombres de las columnas
print(names(df_estudiantes))
print(class(df_estudiantes))
```

# Listas

Son estructuras unidimensionales (solo tienen largo) heterogéneas, es decir, cada uno de sus elementos puede ser de diferente tipo. A diferencia de las estructuras vistas, no se pueden realizar operaciones aritméticas en una lista.

Para saber el número de elementos de la lista se emplea **length**.

## Creación

Se emplea la función **list()**

```python
%%R
lst_mix <- list("dataframe" = df_estudiantes, "matriz" = mat_cols, "vector" = vec_nacionalidades)
lst_mix
```

```python
%%R
print(length(lst_mix))
print(class(lst_mix))
```
